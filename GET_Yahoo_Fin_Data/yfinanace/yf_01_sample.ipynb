download fundamental data for multiple tickers

```python
tickers = ['fb','amzn', 'nflx', 'goog']
tickers
```

```python
tickers = [yf.Ticker(ticker) for ticker in tickers]
tickers
```

```python
dfs = []
for ticker in tickers:

    pnl = dhr.financials
    bs = dhr.balancesheet
    cf = dhr.cashflow

    fs = pd.concat([pnl, bs, cf])

    data = fs.T
    data = data.reset_index()
    data.columns = ['Data', *data.columns[1:]]
    data['Ticker'] = ticker.ticker
    dfs.append(data)
data
```

```python
parser = pd.io.parsers.base_parser.ParserBase({'usecols':None})

for df in dfs:
    df.columns = parser._maybe_dedup_names(df.columns)

df = pd.concat(dfs, ignore_index=True)
df = df.set_index('Ticker', 'Data')
df
```
